```python
import csv
import pprint

def get_rating(reviews):
    if isinstance(reviews, list) and reviews:
        return sum(reviews) // len(reviews)
    elif isinstance(reviews, int):
        return reviews
    else:
        return 0

data = {"Movies": []}

with open("movie_list.txt", "r", encoding="utf-8") as f:
    reader =csv.DictReader(f)
    
    for row in reader:
        data["Movies"].append(dict(row))

for item in data["Movies"]:
    item["rating"] = [int(item["rating"])]
    item["id"]=int(item["id"])
#pprint.pprint(data)
movies=data.get('Movies', [])
#pprint.pprint(movies)
```

```python
while True:
    print("-" * 30)
    print("Best Movie Terminal")
    print("-" * 30)
    print("1. Show Movies")
    print("2. Order Movie")
    print("3. Update Movie")
    print("4. Add Review")
    print("5. Exit")
    print("-" * 30)
    
#Getting input from user
    choice= int(input())
    if choice == 1:
        print('ID\tName\t\tPrice')
        print("-" * 40)
        for movie in movies:
            print(f'{movie.get("id")}\t{movie.get("name")}\t{movie.get("price")}\t{get_rating(movie.get("rating",[]))}')
        print("-" * 40)             
        print("Show Movies")       
    elif choice == 2:
        order_movies=list(map(int, input("What would you like to watch today?").split(",")))
        print('ID\tName\t\tPrice')
        print("-" * 40)
        total_bill=0
        for order_movie in order_movies:
            for movie in movies:
                if movie['id']==(order_movie):
                    print(f'{movie.get("id")}\t{movie.get("name")}\t{movie.get("price")}')
                    total_bill = total_bill + int(movie.get("price",10))
                    break
        print("-" * 40)            
        print(f'\t Total Amount: {total_bill}')
        print("-" * 40)
        print("Order Movies")
    elif choice ==3:
        name= input("Enter movie name")
        movie_price= int(input("Enter price"))
        rating= int(input("Rating"))
        movies.append({
            'id': str(len(movies) + 1),
            'name': name,
            'price': movie_price,
            'rating': rating
        })
        data['Movies']= movies
        with open('movie_list_data.txt', 'w',encoding="utf-8") as f:
            f.write(str(data))
        print("Update Movie")
    elif choice ==4:
        movie_id= int(input("Enter movie id: "))
        rating= int(input("Enter the rating for the movie (1-5): "))
        for i, movie in enumerate(movies):
            if movie["id"]== movie_id:
                movies[i]["rating"].append(rating)
                break
        print("Thank you. Your rating was recorded.")
    else choice == 5:
        with open('movie_list_data.txt', 'w',encoding="utf-8") as f:
            f.write(str(data))
        print("Thank you.")
        break
    
```
